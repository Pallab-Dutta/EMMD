```python
# plot a 1D FEL (non-periodic)

import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
from matplotlib import cm
from matplotlib import rc
import sys

# Read files
inFile = 'FEL_1D_data.npy'         # input file (ue) last column ~ Free energy, other columns ~ corresponding CV value
CV,FE = np.load(inFile).T

# Plotting
plt.figure(dpi=300, figsize=(5,3))
plt.plot(CV,FE,color='crimson',zorder=2)
plt.xlabel('CV')
plt.ylabel('Free Energy (kcal/mol)')

# Formatting the plot (optional)
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(2/4))
ax.xaxis.set_minor_locator(plt.MultipleLocator(2/8))
ax.tick_params(axis='x', colors='black', width=2, length=6, direction='inout')
ax.tick_params(axis='x', which='minor', width=1, length=4, direction='inout')
ax.tick_params(axis='y', colors='black', width=2, length=6, direction='inout')
ax.spines["left"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.spines["top"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
plt.grid(zorder=1)
plt.show()
```

```python
# plot a 2D FEL (non-periodic)

import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
from matplotlib import cm
from matplotlib import rc
import sys

# Read files
inFile = 'FEL_2D_data_a.npy'         # input file (ue) last column ~ Free energy, other columns ~ corresponding CV value
x,y,z = np.load(inFile).T
L = int(np.sqrt(len(x)))
X1=x.reshape(L,L)
X2=y.reshape(L,L)
FE=z.reshape(L,L)

plt.figure(dpi=100)
lbl = [0,1,3,5,10,15,20,25,30]
contour = plt.contour(X1, X2, FE, colors='k', levels=lbl)
contour_filled = plt.contourf(X1, X2, FE, cmap=cm.inferno, levels=lbl)
cbar=plt.colorbar(contour_filled, aspect=20)
plt.xlabel(r'$CV_1$', fontsize=14)
plt.ylabel(r'$CV_2$', fontsize=14)
cbar.set_label('Free Energy (kcal/mol)')
ax = plt.gca()
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
ax.spines["left"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.spines["top"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax1=cbar.ax
ax1.tick_params(axis='y', colors='black')
cbar.outline.set_linewidth(2)
plt.show()
```
